```python
from fuzzywuzzy import process
```

```python
matches = {}
for station_name in manhattan_stations:
    result = process.extract(station_name, stations, limit=2)
    if result[0][1] > 90:
        matches[station_name] = result[0][0]
    else:
        matches[station_name] = ""
```

```python
df1 = pd.DataFrame(columns=['C/A', 'UNIT', 'SCP', 'STATION', 'LINENAME', 'DIVISION', 'ENTRIES', 'EXITS', 'DATETIME'])
d = datetime.date(2019, 12, 7)
for i in range(52,0,-1):
    s = d.strftime('%Y%m%d')
    file_path = path_prefix + s + ".csv"
    df = pd.read_csv(file_path)
    df.rename(columns={'EXITS                                                               ':'EXITS'}, inplace=True)
    df = df[df["DESC"] == "REGULAR"]
    df = df.merge(manhattan_stations_df, on="STATION", how="inner")
    df["DATETIME"] = df.DATE + " " + df.TIME
    df["DATETIME"] = df["DATETIME"].apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%Y %H:%M:%S"))
    df.drop(columns=["DATE", "TIME", "DESC"], inplace=True)
    df.sort_values('DATETIME', ascending=False, inplace=True)
    df1 = pd.concat([df1, df], ignore_index=True, copy=False)
    d = d - timedelta(days=7)
```
